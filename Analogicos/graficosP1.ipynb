```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Soluciones a la ecuacion de schrodinger

def phi1(r): # Solucion para primer nivel energetico
    return np.exp(-r/a)/(np.sqrt(a**3 *np.pi))

def phi2(r): # Solucion para segundo nivel energetico
    return np.exp(-r/(2*a))*(1-r/a)/(2*(np.sqrt(8*a**3 *np.pi)))

def phi3(r): # Solucion para tercer nivel energetico
    return np.exp(-r/(3*a))*(1-(4*r)/(3*a)+(2*r**2)/(9*a**2))/(9*(np.sqrt(3*a**3 *np.pi)))
```

```python
a = 5.29 * 10**(-11) # Radio de Borh [m]

ra = 1500*10**(-12) # radio maximo a graficar [m]

r = np.linspace(0,ra,1000) # barrido de radios de 0 a 1500 [pm]

plt.figure(figsize = (16,10), dpi=300)
plt.rcParams.update({'font.size': 18})
plt.plot(r,4*np.pi*r**2 *phi1(r)**2 *a, label = "n=1")
plt.plot(r,4*np.pi*r**2 *phi2(r)**2 *a, label = "n=2")
plt.plot(r,4*np.pi*r**2 *phi3(r)**2 *a, label = "n=3")
plt.legend()
plt.xlabel("Radio Atomico [m]")
plt.ylabel("Probabilidad de encontrar un electron")
plt.title("Probabilidad $|\psi|^2$ vs Radio atomico")
plt.show()
```
