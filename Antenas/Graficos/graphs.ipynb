```python
# Funcion para graficar los patrones 2D de radiacion

def two_dim_plot(x, y, angles=None):

  plt.figure(figsize=(10,10))
  ax = plt.subplot(111, polar=True)
  ax.set_theta_zero_location('N')
  ax.set_theta_direction('clockwise')
  plt.plot(x, y)
  ax.set_ylim(-20,0)
  ax.set_yticks(np.array([-30, -20, -10, -3, 0, 3]))
  ax.set_yticklabels(['-30dB', '-20dB', '-10dB', '-3dB', '0dB', '3dB'], fontsize=13)
  plt.title(r"Patrón de radiación normalizado [dB]. Plano $\phi=0^o$", size=20)

  if angles:
    ax.set_xticks(np.array(angles)/180*np.pi)
  plt.xticks(fontsize=13)
  plt.show()
```
