```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
#import 
sparametro = pd.read_csv("https://raw.githubusercontent.com/aitkenbug/shiny-funicular/main/Antenas/Graficos/data/spiders11.csv")
hfield = pd.read_csv("C:\\Users\\kabuto\\Documents\\Github\\shiny-funicular\\Antenas\\Graficos\\data\\Hfieldspider.csv")
efield = pd.read_csv("https://raw.githubusercontent.com/aitkenbug/shiny-funicular/main/Antenas/Graficos/data/Efieldspider.csv")
```

```python
y1 = sparametro["S11-S-Real"].to_numpy()
y2 = sparametro["S11-S-Imaginary"].to_numpy()
y = 20*np.log10(np.abs(y1,y2))
x = sparametro["Frequency"]
plt.figure(figsize=(20,10), dpi=120)
plt.rcParams.update({'font.size': 18})
plt.plot(x, y)
# plt.yticks(np.arange(-22,2,2))
plt.title("Barrido de frecuencia 200MHz-2GHz")
plt.xlabel("Freq [GHz]")
plt.ylabel("dB(S(1,1))")
plt.legend(['150mm'])
plt.grid()
plt.savefig("parametro-s-aranha.png")
```

```python
def to_rad(deg):
    rad1 = []
    for i in deg:
        rad1.append(i*0.017453)
    return rad1
```

```python
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
ax.plot(to_rad(hfield['grad']), hfield['dB'], color='r', linewidth=3)
ax.grid(True)
ax.set_theta_zero_location('N')
# ax.set_yticks(np.array([-60,-50,-40,-30,-20,-10,0]))
# ax.set_yticklabels(["-60dB","-50dB","-40dB","-30dB","-20dB","-10dB","0dB"])
ax.set_title("Patron de radiacin Antena ADS-B", va='bottom')
plt.savefig('radiaton-pattern-h.png')
plt.show()
```

```python
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
ax.plot(to_rad(efield['grad']), efield['dB'], color='r', linewidth=3)
ax.grid(True)
ax.set_theta_zero_location('N')
# ax.set_yticks(np.array([-60,-50,-40,-30,-20,-10,0]))
# ax.set_yticklabels(["-60dB","-50dB","-40dB","-30dB","-20dB","-10dB","0dB"])
ax.set_title("Patron de radiacin Antena ADS-B", va='bottom')
plt.savefig('radiaton-pattern-e.png')
plt.show()
```
