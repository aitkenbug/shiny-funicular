```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
#import 
sparametro = pd.read_csv("C:\\Users\\kabuto\\Documents\\Github\\shiny-funicular\\Antenas\\Graficos\\data\\spiders11.csv")
sparametro
```

```python
sparametro["Frequency"]
```

```python
y1 = sparametro["S11-S-Real"].to_numpy()
y2 = sparametro["S11-S-Imaginary"].to_numpy()
y = 20*np.log10(np.abs(y1,y2))
x = sparametro["Frequency"]
plt.figure(figsize=(20,10), dpi=120)
plt.rcParams.update({'font.size': 18})
plt.plot(x, y)
# plt.yticks(np.arange(-22,2,2))
plt.title("Barrido de frecuencia 200MHz-2GHz")
plt.xlabel("Freq [GHz]")
plt.ylabel("dB(S(1,1))")
plt.legend(['150mm'])
plt.grid()
plt.savefig("Barrido de frecuencia 200MHz-2GHz.png")
```
